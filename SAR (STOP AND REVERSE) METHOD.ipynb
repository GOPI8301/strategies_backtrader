```python
import backtrader as bt
import numpy as np
import datetime as dt
from datetime import datetime

class TimeSARexample(bt.Strategy):
    params = (
            ('Printlog', True),
        )
    
    def log(self, txt, dt=None):
        print(txt)
        
    def __init__(self):
        self.data.close = self.datas[0].close
        self.sma = bt.indicators.SimpleMovingAverage(self.datas[0],period=self.params.malookback)
        self.hilodiff = self.datas[0].high - self.datas[0].low
        self.sma20 = bt.indicators.SMA(self.datas[0], period=20)
        self.sma50 = bt.indicators.SMA(self.datas[0], period=50)
        self.ma_cd = MACD(self.datas[0])
        bt.indicators.WeightedMovingAverage(self.datas[0], period=20, subplot=True)
        bt.indicators.StochasticSlow(self.datas[0])
        bt.indicators.MACDHisto(self.datas[0])
        self.rsi = bt.indicators.RSI(self.datas[0])
        bt.indicators.ATR(self.datas[0], plot=False)
        self.stoporder = None
        
    def notify_order(self, order):
        if order.status in [Order.Submitted, Order.Accepted]:
            return
        if Order.status in [Order.Completed]:
            if order.isbuy():
                pass
            elif order.issell():
                pass
            self.bar_executed = len(self)
        elif order.status in[order.Canceled, order.Margin, order.Rejected]:
            pass
        def next(self):
            if not self.position:
                if self.sma20[0] > self.sma50[0]:  
                    print("Entering Buy Order")
                    self.order = self.buy()
                elif self.sma20[0] < self.sma50[0]:  
                    print("Entering Sell Order")
                    self.order = self.sell()
            else:
                if self.order.isbuy():
                    if self.sma20[0] < self.sma50[0]:
                        print("Closing buy order & entring Sell")   # If 20 period MA cross above 50 period MA (close short & go long)
                        self.Close()
                        self.order = self.sell()
                elif self.order.issell():
                    if self.sma20[0] > self.sma50[0]:
                        print("Closing sell order & entering Buy")  # If 50 period MA cross above 20 period MA (20 ma cross below 50 MA)(close long & go short)
                        self.close()
                        self.order = self.buy()
        def stop(self):
            print("Final portfolio value: ", self.broker.getvalue())
            
if __name__=='__main__':
    cerebro = bt.Cerebro(stdstats = True)
    comminfo = CommInfo(
            commission = 0.001,
            mult = 75,
            margin = 100000
            )
    
    cerebro.broker.addcommissioninfo(comminfo)
    
    datapath = "../Data/CLEANEDCDATA/Nifty_F1_5m_data_clean.csv"
    data = bt.feeds.GenericCSVData(
        dataname=datapath,
        fromdate=datetime.datetime(2019,1,1),
        todate=datetime.datetime(2021,7,21),
        datetime= 0,
        timeframe = bt.TimeFrame.Days,
        compression = 1,
        dtformat = ('%Y-%m-%d-%H-%M'),                 
        Open = 1,
        high = 2,
        low = 3,
        close = 4,
        volume = None,
        openinterest = None,
        reverse=False,
        header=0)

    cerebro.adddata(data)
    cerebro.addsizer(bt.sizers.FixedSize, stake=1)                  
    cerebro.broker.setcommission(commission=0.001)                     
    cerebro.broker.setcash(100000.0)
    print('Starting Portfolio Value: ', cerebro.broker.getvalue())
    cerebro.run()
    print('Final Portfolio Value: ', cerebro.broker.getvalue())
                
```
